```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
data = pd.read_csv('data/sortby_MAC_Time.csv', header=[0,1], index_col=0)
```

```python
dic = {}
for col1, col2 in data.columns:
    if col1!= 'l1':
        dic[col1] = ''
        strs = col1.split('.')
        dic[col1] = strs[0]

data.rename(columns = dic, inplace = True)
```

```python
access = data[[('AP','Map Location'), ('Client','MAC Address'),('Client','Endpoint Type'), ('Disassociation Time','Unnamed: 23_level_1' ),('Association Time', 'Unnamed: 24_level_1')]]
```

```python
# 1. 장전캠퍼스만
```

```python
access = data[data[('AP','Map Location')].str.startswith('Pusan')]
```

```python
# size 칼럼 추가
size = access.groupby(('Client','MAC Address')).size()
```

```python
# access[('Client','size')] = 1
access[('Client','size')] = size[access['Client','MAC Address']].values
```

```python
# 건물 이름 칼럼 추가
```

```python
# 건물 이름 분리하기
sp = access[('AP','Map Location')].str.split('>')
```

```python
lst=[]
for loc in sp:
    lst.append(loc[1].strip())
```

```python
access[('AP','Building')] = lst
```

```python
data_col = [('AP','Building'), ('Client','MAC Address'),('Client','Endpoint Type'), ('Disassociation Time','Unnamed: 23_level_1' ),('Association Time', 'Unnamed: 24_level_1')]
```

```python
userlist = list(freq_data[('Client','MAC Address')].unique())
```

```python
#1. 사용자별로 돌아다니는 건물 튜플로 정리
```

```python
user_use_building ={}
for mac_num in userlist:
    user = freq_data[freq_data[('Client','MAC Address')] == mac_num][data_col]
    lst = list(user[('AP','Building')].unique())
    user_use_building[mac_num] = lst
```

```python
# 3. 사용자별로 학교에 있는 시간대 정리 - 그래프(날짜-시간 그래프)
# 4. 사용자의 날짜별 접속 횟수 xxx 나중에
# 5. 사용자의 날짜별 누적 접속 시간 xxxx 나중에
```

```python
# 3. 사용자별로 학교에 있는 시간대 정리
```

```python
user_come_out ={}
pd.set_option('display.max_rows', 100)

for mac_num in userlist:
    # user 정하기
    user = freq_data[freq_data[('Client','MAC Address')] == mac_num][data_col]
    
        
    # 출근시간
    come = user[('Association Time','Unnamed: 24_level_1')].str.split(' ')
    # 퇴근시간
    out = user[('Disassociation Time','Unnamed: 23_level_1')].str.split(' ')
    
    
    dic_days={}
    come_times=[]
    one_day = []
    out_times=[]
    
    for c,o in zip(come, out):
        store_day = ''
        temp='00:00:00'
        # 날짜 저장
        
        # 적었던 날이 아니면
        if c[0] not in one_day:
            store_day = c[0]
            
            
            if c[1] > '05:00:00':
                lst = []
                lst.append(store_day)
                lst.append(c[1])
                one_day.append(c[0])
                come_times.append(lst)
                    
            
            #시간 - 한번 적었던 날이어도 괜찮다
            if o[1] > temp:
                temp = o[1]
                lst = []
                lst.append(store_day)
                lst.append(temp)
                out_times.append(lst)
                    
                
            dic_days['come'] = come_times
            dic_days['out'] = out_times
            
    user_come_out[mac_num] = dic_days  
```
