```
dict = set(data['AP.3'])

loc = []

for d in dict :
    cur = d.split('>')
    if len(cur) >= 2 :
        loc.append(cur[1])

        
loc.sort()
loc = set(loc)

ret = []

for d in loc : 
    if d.find('(') != -1 :        
        ret.append(d)

ret.remove(' BoYuk(New) ')

#ret에 바꿔야할 건물 저장되어 있음


for dd in ret:
    curdata = dd.split(' ')
    cdata = curdata[1] # cdata[1] 은 Haksang (708) 에서 Haksang 의미    
    
    for d in data['AP.3'][ : 10]:                        
        cur = d.split('>') 
        if len(cur) < 2: continue
            
        realdata = cur[1].split(' ')
        rdata = realdata[1] #rdata 가 데이터에서 가운데 건물 의미한느 값
        
        
        ans = str(cur[0])+'>'+str(dd)+'>'+str(cur[2]) #cdata rdata가 같으면 ans로 바꿔주면 됨
        if cdata == rdata:            
            print(ans)
```

```
a = {}

for b in ret :
  b = b.split(' ')
  if b[1] not in a :
    a[b[1]] = b[2]

```

```
def func(row):
  st = row.split(' ')
  if len(st) < 5 : return row
  if st[2] not in a : return row
  if st[3].startswith('('):
    return " ".join(st)
  else :
    st.insert(3, a[st[2]])
    return " ".join(st)
  
```

```
temp = data
temp['AP.3'] = temp['AP.3'].apply(lambda x : func(x))
```
