# Android / Apple / Workstation 의 개수 그래프

```python
plt.rcParams["figure.figsize"] = (10,10)
df = data['Client.4'].value_counts().sort_values()

type(df)

x = dict()
x['Laptop'] = 0
x['SmartPhone'] = 0
x['Pad'] = 0
x['Else'] = 0

for index, value in df.items():
    if 'Workstation' in index:
        x['Laptop'] = x['Laptop']+value
    elif 'Android' in index or 'Phone' in index or 'Samsung' in index:
        x['SmartPhone'] = x['SmartPhone']+value
    elif 'Pad' in index:
        x['Pad'] = x['Pad']+value
    else:
        x['Else'] = x['Else']+value

lists = sorted(x.items(), key=(lambda x:x[1]), reverse=True)
print(lists)
a, b = zip(*lists)      

plt.plot(a,b)
plt.text('Else', 6000000, '               Total number=9775563')
plt.xlabel('device')
plt.ylabel('number')
plt.title("Number of appearances by device")
plt.show()
```

# 제도관 층별 데이터 읽기

```python
y_column = 'Client.4'
data1 = pd.read_csv('data\clients_11S-201-1-3.csv')

y_true_label_encoder = preprocessing.LabelEncoder()
y_true_encoded       = y_true_label_encoder.fit_transform(data1[y_column]) # EndPoint Type 문자열이므로 라벨링

index = 1
#RSSI str -> int
for i in data1['Utilization.6'][1:]:
    data1['Utilization.6'][index] = int(i)
    index = index+1
    
data1['endlabel'] = y_true_encoded

data2 = pd.read_csv('data\clients_11S-201-1-4.csv')

y_true_label_encoder = preprocessing.LabelEncoder()
y_true_encoded       = y_true_label_encoder.fit_transform(data2[y_column]) # EndPoint Type 문자열이므로 라벨링

index = 1
#RSSI str -> int
for i in data2['Utilization.6'][1:]:
    data2['Utilization.6'][index] = int(i)
    index = index+1
    
data2['endlabel'] = y_true_encoded

data3 = pd.read_csv('data\clients_11S-201-2-5.csv')

y_true_label_encoder = preprocessing.LabelEncoder()
y_true_encoded       = y_true_label_encoder.fit_transform(data3[y_column]) # EndPoint Type 문자열이므로 라벨링

index = 1
#RSSI str -> int
for i in data3['Utilization.6'][1:]:
    data3['Utilization.6'][index] = int(i)
    index = index+1
    
data3['endlabel'] = y_true_encoded

print(len(data1))
print(len(data2))
print(len(data3))
```

#  RSSI 와 EndPoint Type으로 K-means Clustering 진행

```python
#x_columns = ['Utilization.8', 'endlabel']
x_columns = ['Utilization.6','endlabel']

model1 = KMeans(n_clusters=4, random_state=0, n_init=100)  # k개 클러스터로 모델 생성
cluster_labels1 = model1.fit_predict(data1[x_columns][1:]) # X컬럼으로 지정된 필드값으로 피팅

zero = [0 for _ in range(1850)]
data1['cluster'] = np.array(np.array(zero))
data1['cluster'][1:] = cluster_labels1

model2 = KMeans(n_clusters=4, random_state=0, n_init=100)  # k개 클러스터로 모델 생성
cluster_labels2 = model2.fit_predict(data2[x_columns][1:]) # X컬럼으로 지정된 필드값으로 피팅

zero = [0 for _ in range(2341)]
data2['cluster'] = np.array(np.array(zero))
data2['cluster'][1:] = cluster_labels2

model3 = KMeans(n_clusters=4, random_state=0, n_init=100)  # k개 클러스터로 모델 생성
cluster_labels3 = model3.fit_predict(data3[x_columns][1:]) # X컬럼으로 지정된 필드값으로 피팅

zero = [0 for _ in range(11450)]
data3['cluster'] = np.array(np.array(zero))
data3['cluster'][1:] = cluster_labels3
```

```python
g1 = sns.lmplot('Utilization.6', 'endlabel', data=data1[1:], fit_reg = False, 
         scatter_kws={"s":100}, hue='cluster')

g1.set_axis_labels("RSSI", "EndPoint Type")
plt.title('201-3', size=20)

g2 = sns.lmplot('Utilization.6', 'endlabel', data=data2[1:], fit_reg = False, 
         scatter_kws={"s":100}, hue='cluster')

g2.set_axis_labels("RSSI", "EndPoint Type")  
plt.title('201-4', size=20)

g3 = sns.lmplot('Utilization.6', 'endlabel', data=data3[1:], fit_reg = False, 
         scatter_kws={"s":100}, hue='cluster')

g3.set_axis_labels("RSSI", "EndPoint Type")  
plt.title('201-5', size=20)
```
